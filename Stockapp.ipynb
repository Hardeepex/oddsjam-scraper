```python
from selenium import webdriver
from time import sleep
from random import seed
from random import randint
from selenium.webdriver.common.by import By
import smtplib, ssl
```

```python
def login():
    browser = webdriver.ChromiumEdge()
    browser.get('https://thefinancialcloud.memberful.com/auth/sign_in')
    browser.implicitly_wait(5)

    ## Link to signup
    signin_link = browser.find_element(By.CLASS_NAME, "underline")        
    signin_link.click()
    browser.implicitly_wait(5)        
    # Put log in info
    mail_input = browser.find_element(By.ID, "session_email")        
    mail_input.send_keys("REDACTED@gmail.com") # Add key file
    password_input = browser.find_element(By.ID, "session_password") 
    password_input.send_keys("REDACTED") # Add key file
    ## Click to log in
    log_in = browser.find_element(By.CLASS_NAME, "button")
    ## Uses JavaScript arguments to click        
    browser.execute_script("arguments[0].click();", log_in)
    browser.implicitly_wait(10)
    ## click on flowchart
    flow_chart = browser.find_element(By.CSS_SELECTOR,"a[href='https://thefinancial.cloud/flow-alerts/']")
    ## Uses JavaScript arguments to click
    browser.execute_script("arguments[0].click();", flow_chart)
    return browser
```

```python
def getFlowChart(browser):
    # locate the table with the flow_chart
    table_id = browser.find_element(By.ID, 'alertsTableBody')
    # get all of the rows in the table
    rows = table_id.find_elements(By.TAG_NAME, "tr") 
    # collec the data from flow_chart
    flowInfo = [] 
    for row in rows:
        # colunm on each row
        col = row.find_elements(By.TAG_NAME, "td")
        # I cant only calls and rapid signal
        # col[3] = CALL/PUt and col[1] = signal type
        signals = ["Rapid", "Whale","Oracle"]
        if col[3].text == "CALL" and col[1].text in signals :
        
            # Get time       
            time = col[0].text 
        
            # get stock name 
            stock = col[2].text
            # get TFC score
            score = col[9].text
            # append info
            flowInfo.append([stock,time,score])
    return flowInfo
```

```python
## log in to TFC and go to flowcharts
browser = login()
sender_email = "REDACTED@gmail.com"
receiver_email = "REDACTED@gmail.com"
```

```python
def formatChart(flowInfo):
    text = ""
    for chart in flowInfo:
        text += chart[0] + "-->" + chart[2] +"  "+ chart[1]+"\n" 
    return text
```

```python
## records the count of charts
chartCount = 0
```

```python
while True:
    flowInfo = getFlowChart(browser)
    if len(flowInfo) > chartCount:
        chartCount = len(flowInfo)
        
        text = formatChart(flowInfo)
        message = "Subject: Flowchart \n\n"+text
        print(message)
        port = 465  # For SSL
        password = "REDACTED" # add keyfile
        # Create a secure SSL context
        context = ssl.create_default_context()
    
        with smtplib.SMTP_SSL("smtp.gmail.com", port, context=context) as server:
            server.login("REDACTED@gmail.com", password) 
            server.sendmail(sender_email, receiver_email, message)
    else:
        print("No new charts")
    sleep(20)
```
